```python
import pandas as pd

df = pd.read_csv("owid-covid-data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[['date', 'location', 'total_cases', 'new_cases', 'total_deaths', 'population']]
countries = ['India', 'United States', 'Brazil']
df = df[df['location'].isin(countries)]
```

```python
df['location'].unique()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['total_cases', 'new_cases'])
```

```python
df = df.sort_values(by=['location', 'date'])
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
df.info()
```

```python
df['cases_per_million'] = df['total_cases'] / df['population'] * 1_000_000
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
for country in df['location'].unique():
    data = df[df['location'] == country]
    plt.plot(data['date'], data['total_cases'], label=country)

plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.title("COVID-19 Total Cases Over Time")
plt.legend()
plt.show()
```

```python
for country in df['location'].unique():
    data = df[df['location'] == country]
    plt.plot(data['date'], data['new_cases'], label=country)

plt.xlabel("Date")
plt.ylabel("New Cases")
plt.title("Daily New COVID-19 Cases")
plt.legend()
plt.show()
```

```python
for country in df['location'].unique():
    data = df[df['location'] == country]
    plt.plot(data['date'], data['total_deaths'], label=country)

plt.xlabel("Date")
plt.ylabel("Total Deaths")
plt.title("COVID-19 Death Trends")
plt.legend()
plt.show()
```

```python
for country in df['location'].unique():
    data = df[df['location'] == country]
    plt.plot(data['date'], data['cases_per_million'], label=country)

plt.xlabel("Date")
plt.ylabel("Cases per Million")
plt.title("Fair Comparison (Cases per Million)")
plt.legend()
plt.show()
```

```python
df['MA7'] = df.groupby('location')['new_cases'].transform(lambda x: x.rolling(7).mean())
```

```python
for country in df['location'].unique():
    data = df[df['location'] == country]
    plt.plot(data['date'], data['MA7'], label=country)

plt.title("7-Day Moving Average of New Cases")
plt.legend()
plt.show()
```
